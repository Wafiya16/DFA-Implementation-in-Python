```python
class DFA:

    def __init__(self, states, alphabet, transition_function, start_state, accept_states):
        self.states = states
        self.alphabet = alphabet
        self.transition_function = transition_function
        self.start_state = start_state
        self.accept_states = accept_states
        self.current_state = start_state
        self.input_symbols = []
        self.accepted = False

    def check_input(self, input_value):
        for i in input_value:
            if i not in alphabet:
                return False
        return True

    def check_state(self, state):
        if state in states:
            return True
        else:
            return False

    def check_transition(self, state, input_value):
        if (state, input_value) in transition_function.keys():
            return True
        else:
            return False

    def run(self, input_list):
        self.current_state = self.start_state
        self.input_symbols = []
        for i in input_list:
            if self.check_state(self.current_state) and self.check_transition(self.current_state, i):
                self.current_state = self.transition_function[(self.current_state, i)]
                self.input_symbols.append(i)
            else:
                raise Exception("Invalid Input: No Transition Function")
        self.accepted = self.current_state in self.accept_states
        return self.accepted
```

DFA Example 1


```python
from IPython.display import Image
Image('dfa1.png')
```

```python
states = {"q0", "q1", "q2", "q3"}
alphabet = {"0", "1"}
transition_function = {
    ("q0", "0"): "q1",
    ("q0", "1"): "q3",
    ("q1", "0"): "q0",
    ("q1", "1"): "q2",
    ("q2", "0"): "q3",
    ("q2", "1"): "q1",
    ("q3", "0"): "q2",
    ("q3", "1"): "q0"
}
start_state = "q0"
accept_states = {"q0"}

dfa1 = DFA(states, alphabet, transition_function, start_state, accept_states)

with open('DFA1_input.txt') as f:
    input_value = f.read().splitlines()
print("DFA1 Input: ", input_value)
for i in input_value:
    if dfa1.check_input(i):
        result = dfa1.run(i)
        if result:
            print(i, " : Accepted")
        else:
            print(i, " : Rejected")
    else:
        print(i, " : Invalid Input")
```

```python
states = {"q0", "q1", "q2", "q3"}
alphabet = {"a", "b", "c"}
start_state = "q0"
accept_states = {"q2"}
transition_function = {
    ("q0", "a"): "q2",
    ("q0", "b"): "q2",
    ("q0", "c"): "q1",
    ("q1", "a"): "q3",
    ("q1", "b"): "q2",
    ("q1", "c"): "q2",
    ("q2", "a"): "q3",
    ("q2", "b"): "q3",
    ("q2", "c"): "q3",
    ("q3", "a"): "q3",
    ("q3", "b"): "q3",
    ("q3", "c"): "q3"
}
dfa3 = DFA(states, alphabet, transition_function, start_state, accept_states)

with open('DFA3_input.txt') as f:
    input_value = f.read().splitlines()
print("DFA3 Input: ", input_value)
for i in input_value:
    if dfa3.check_input(i):
        result = dfa3.run(i)
        if result:
            print(i, " : Accepted")
        else:
            print(i, " : Rejected")
    else:
        print(i, " : Invalid Input")
```
